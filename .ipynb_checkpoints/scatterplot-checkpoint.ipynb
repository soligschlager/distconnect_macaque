```python
% matplotlib inline

import numpy as np, nibabel as nib
import utils.plotting_lighterbg as plot
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from scipy.stats import spearmanr, pearsonr
import pandas as pd
import statsmodels.formula.api as smf
from adjustText import adjust_text
mpl.rcParams['svg.fonttype'] = 'none'

sns.set_style('white')
#sns.set_context('talk')
sns.set_context('paper')

df_area = pd.read_pickle('./data/df_revrequest_outlcomp.pkl')
```

```python
#fig, ax = plt.subplots()
#fig.set_size_inches(8,8)

g = sns.jointplot(x="primdist", y="meandist", data=df_area, kind='reg', 
                  color= "#533cc6", 
                  xlim=(0,30), ylim=(0,34),
                  stat_func=None)

_ = plt.xticks([0,10,20,30])
_ = plt.yticks([0,10,20,30])

texts = [plt.text(df_area.primdist[i], df_area.meandist[i], '%s' % df_area.area[i], ha='center', va='center') for i in df_area.index]
#adjust_text(texts, arrowprops=dict(arrowstyle='-', color='grey'))

adjust_text(texts, arrowprops=dict(arrowstyle="-", color='grey', lw=0.5), 
            autoalign=None, only_move={'points':'y', 'text':'y'}, force_text=(0.5, 0.25), expand_text=(1.5, 1.2))#, autoalign='')
            # # We want them to be quite compact, so reducing expansion makes sense
             #With default forces it takes a very long time to converge, but higher values still produce very nice output

#g.savefig('./scatterplot.pdf', dpi=300)
g.savefig('./scatterplot.svg', dpi=300)
```
