## functions

```python
# calculate distance from these

def get_dist(nodes, df, gdist_mat):
    
    min_dists = list()
   
    for idx in df.index:
        inj_nodes = df['injection sites'].loc[idx]    
        dists = [gdist_mat[inj_nodes,:][:,node].mean() for node in nodes]
        min_dists.append(np.min(np.array(dists)))
    
    return min_dists
```

```python
# glm

def run_glm(df):
    
    full = smf.ols(formula='incoming ~ location \
                                     + region_size \
                                     + anchor_dist', data=df).fit()
    if full.params[3] > 0:
        return full.rsquared
    if full.params[3] < 0:
        return -full.rsquared
```

```python
def find_new_nodes(rois_orig, gdist_mat, cort, rlt='pos', min_dist=0, max_dist=1):

    # define patches around random seed nodes
    
    nodes_prox = np.where(((gdist_mat[rois_orig] <= max_dist) 
                         & (gdist_mat[rois_orig] >= min_dist)) 
                         | (gdist_mat[rois_orig] == 0))[1]

    nodes_prox_grp = np.where(((gdist_mat[rois_orig] <= max_dist)
                             & (gdist_mat[rois_orig] >= min_dist)) 
                             | (gdist_mat[rois_orig] == 0))[0]
    

    nodes_prox_grp = np.array([nodes_prox_grp[i] for i, node in enumerate(nodes_prox) 
                               if node in cort])
    nodes_prox = np.array([node for node in nodes_prox if node in cort])

    
    
    # iterate over patches and find local minima for each
    
    rois_new = list()

    for i, _ in enumerate(rois_orig):

        rois = copy.copy(rois_orig)
        patch = nodes_prox[np.where(nodes_prox_grp==i)]
      
        rsqrs = list()

        for n, node in enumerate(patch):
            
            rois[i] = node
            df.anchor_dist = get_dist(rois, df, gdist_mat)
            rsqrs.append(run_glm(df))
        
        if rlt == 'pos':
            rois_new.append(patch[np.argmax(rsqrs)])
        if rlt == 'neg':
            rois_new.append(patch[np.argmin(rsqrs)])
        
    return np.array(rois_new)
```

```python
def find_minima(count, nodes, cort, gdist_mat, return_dict, set_size=3, min_dist=0, max_dist=1):

    #nodes = np.array(random.sample(cort, set_size))
    
    log = [np.zeros((nodes.shape))]

    while not all(nodes == log[-1]):

        if any([all(x == nodes) for x in log]):
            return None
            break

        log.append(nodes)
        nodes = find_new_nodes(nodes, gdist_mat, cort, min_dist=min_dist, max_dist=max_dist)

    else:
        return_dict[count] = nodes
```

```python
def find_extrema(count, nodes, cort, gdist_mat, return_dict, 
                 minmax='minima', set_size=3, min_dist=0, max_dist=1):

    if minmax == 'minima':
        rlt = 'pos'
    else:
        rlt = 'neg'
    
    log = [np.zeros((nodes.shape))]

    while not all(nodes == log[-1]):

        if any([all(x == nodes) for x in log]):
            return None
            break

        log.append(nodes)
        nodes = find_new_nodes(nodes, gdist_mat, cort, 
                               rlt=rlt, min_dist=min_dist, max_dist=max_dist)

    else:
        return_dict[count] = nodes
```

## search procedure

```python
# set params
n_pll = 10
iterations = 5000
n_rois = range(3,6)

# start
manager = multiprocessing.Manager() 
return_dict_min = manager.dict()
return_dict_max = manager.dict()
jobs = []


count = 0
while count < iterations:
    
    for i in range(n_pll):
    
        rand_nodes = np.random.choice(cort, np.random.choice(n_rois))

        p_min = multiprocessing.Process(target = find_extrema,
                                        args = (count+i, rand_nodes, cort, gdist_mat, 
                                                return_dict_min, 'minima'))
        
        p_max = multiprocessing.Process(target = find_extrema,
                                        args = (count+i, rand_nodes, cort, gdist_mat, 
                                                return_dict_max, 'maxima'))

        jobs.append(p_min)
        p_min.start()
        jobs.append(p_max)
        p_max.start()
    p_min.join() # move two right for serial, move to left for complete parallel comp
    p_max.join()
    count += n_pll
```

```python
# IMMEDIATELY SAVE RETURN DICT INTO LIST
L_min = list()
L_max = list()

for l in return_dict_min.values():
    for x in l:
        L_min.append(x)
        
for l in return_dict_max.values():
    for x in l:
        L_max.append(x)
```

```python
data_max = np.zeros(gdist_mat.shape[0])
for node in L_max:
    data_max[node] += 1
    
data_min = np.zeros(gdist_mat.shape[0])
for node in L_min:
    data_min[node] += 1
    
data_maxmin = np.zeros(gdist_mat.shape[0])
for node in L_max:
    data_maxmin[node] += 1
for node in L_min:
    data_maxmin[node] -= 1
```
