```python
% matplotlib inline

import pandas as pd, numpy as np, nibabel as nib
import seaborn as sns
import matplotlib.pyplot as plt
import utils.plotting_lighterbg as plot

sns.set_style('white')
sns.set_context('talk')
```

```python
#data_dir = '/Users/sabine/workspace/distconnect_macaque/data'
yerkes_dir ='/Users/sabine/data/Yerkes19'


# macaque surface
f_surf = '%s/MacaqueYerkes19.L.inflated.32k_fs_LR.surf.gii' % yerkes_dir
coords = nib.load(f_surf).darrays[0].data
faces = nib.load(f_surf).darrays[1].data
surf = tuple((coords, faces))
f_sulc = '%s/MacaqueYerkes19.sulc.32k_fs_LR.dscalar.nii' % yerkes_dir
sulc = nib.load(f_sulc).get_data().squeeze()[:32492]

# labels
mwall = np.load('%s/medial_wall.lh.label.npy' % yerkes_dir)
cort = np.array([node for node in range(32492) if node not in mwall])
areas_f = '%s/MarkovCC12_M132_91-area.32k_fs_LR.dlabel.nii' % yerkes_dir
areas_annot = nib.load(areas_f).get_data().squeeze()[:28261]
areas_key = np.load('%s/M132_key.npy' % yerkes_dir)

# distances
gdist_mat = np.load('%s/yerkes19_gdist_node-by-node.npy' % yerkes_dir)
```

# Distance from primary cortex
### with primary and excluded areas outlined

```python
prim_areas = ['V1', 'F1', '3', 'Core']
primary_keys = [np.where(areas_key == area)[0][0] for area in prim_areas] 
primary_nodes_nested = [np.where(areas_annot == key)[0] for key in primary_keys]
primary_nodes = cort[sorted([node for area in primary_nodes_nested for node in area])]

distprim = np.min(gdist_mat[np.array(primary_nodes)], axis=0)
distprim[np.isinf(distprim)] = 0

labels = [cort[np.where(areas_annot == key)[0]] for key in primary_keys]

excl_areas = ['10', '11', '13', '14', '24c', '25', '32', '46d', '46v', '9', 'F6', 'OPAI']
excl_keys = [np.where(areas_key == area)[0][0] for area in excl_areas]
excl_labels = cort[np.isin(areas_annot, excl_keys)]

labels.append(excl_labels)

for azim in [0, 180]:
    img = plot.plot_surf_stat_map(coords, faces,
                                  distprim, mask=cort,
                                  labels = labels,
                                  label_cpal = ['green' for _ in prim_areas] + ['aqua blue'],
                                  azim=azim,
                                  sulc=sulc, bg_map=sulc, 
                                  bg_on_stat=True, cmap='CMRmap')
    
    plt.savefig('/Users/sabine/Downloads/gdist_from_prim_%s' % (azim), dpi=300)
```

# Original distance from primary

```python
temp = (np.where((areas_key == 'V1') | (areas_key == 'F1') | (areas_key == '3') | (areas_key == 'Core')))[0]
primary_nodes = cort[np.where((areas_annot == temp[0]) |
                              (areas_annot == temp[1]) |
                              (areas_annot == temp[2]) |
                              (areas_annot == temp[3]))[0]]

distprim = np.min(gdist_mat[primary_nodes], axis=0)
distprim[np.isinf(distprim)] = 0

labels = (cort[np.where(areas_annot == temp[0])[0]], 
          cort[np.where(areas_annot == temp[1])[0]], 
          cort[np.where(areas_annot == temp[2])[0]],
          cort[np.where(areas_annot == temp[3])[0]])

for azim in [0, 180]:
    img = plot.plot_surf_stat_map(coords, faces,
                                  distprim, mask=cort,
                                  labels = labels, azim=azim,
                                  bg_map=sulc, bg_on_stat=True, cmap='CMRmap')
    
    #plt.savefig('/nobackup/hunte1/sabine/tractdist/figures_ms/images/primdist_yerkes_%s' % (azim), dpi=300)
```
